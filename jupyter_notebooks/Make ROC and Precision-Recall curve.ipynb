# This makes a roc and precision-recall curve for Experiment 3

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
from os.path import join
```

# Load Data for Experiment 1 method
## Experiment 1 method uses training data that comes from testing of Avenue dataset

```python
plt.style.use('classic')
```

```python
exp_1_loc = '/home/akanu/git/anomalous_pred/plots/experiment_3/Avenue/using_exp_1_method'
frames =20
```

```python
nc_bin_exp_1 = ['seed{:d}_Dense_5_Drop_5_exp_1', 'xywh','avenue', frames, 0.500]
```

```python
data_exp_1_fpr_tpr = {}
for i in (23,24,35,81):
    nc_bin_exp_1[0] = 'seed{}_Dense_5_Drop_5_exp_1'
    nc_bin_exp_1[0]=nc_bin_exp_1[0].format(i)
    temp = pd.read_csv(join(exp_1_loc, 
                        'fpr_tpr_exp_1_{}_{}_{}_{}_{}.csv'.format(
                                                    nc_bin_exp_1[0],
                                                    nc_bin_exp_1[1],
                                                    nc_bin_exp_1[2],
                                                    nc_bin_exp_1[3],
                                                    nc_bin_exp_1[4])))
    data_exp_1_fpr_tpr[i] = temp
```

```python
fig, ax= plt.subplots(1,1)
for i in data_exp_1_fpr_tpr.keys():
    ax.plot(data_exp_1_fpr_tpr[i]['fpr'],
            data_exp_1_fpr_tpr[i]['tpr'],
            label='seed {}'.format(i))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.title('Experiment 1 method')
fig.savefig(join(exp_1_loc, 'roc_exp_1_method.jpg'))
```

```python
data_exp_1_pre_recall = {}
for i in (23,24,35,81):
    nc_bin_exp_1[0] = 'seed{}_Dense_5_Drop_5_exp_1'
    nc_bin_exp_1[0]=nc_bin_exp_1[0].format(i)
    temp =pd.read_csv(join(exp_1_loc,
                           'prec_recall_thre_exp_1_{}_{}_{}_{}_{}.csv'.format(
                                                    nc_bin_exp_1[0],
                                                    nc_bin_exp_1[1],
                                                    nc_bin_exp_1[2],
                                                    nc_bin_exp_1[3],
                                                    nc_bin_exp_1[4])))
    data_exp_1_pre_recall[i] = temp
```

```python
fig1, ax1= plt.subplots(1,1)
# line = ['--', '--.', '-', '-.']
for i in data_exp_1_pre_recall.keys():
    
    ax1.plot(data_exp_1_pre_recall[i]['recall'],
            data_exp_1_pre_recall[i]['precision'], 
            label='seed {}'.format(i))
ax1.set_xlabel('Recall')
ax1.set_ylabel('Precision')
plt.legend(loc='best')
plt.title('Experiment 1 method')
fig1.savefig(join(exp_1_loc, 'pre_recall_exp_1_method.jpg'))
```
